## Parsing

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text,'html.parser')
soup
```

```python
souplettesBody = soup.find('body')
souplettesBody
```

```python
finded = souplettesBody.findAll('span',{"class":"Head"})
finded
```

```python
for v in finded:
    spans = v.findAll('span')
    for span in spans:
        if span.has_attr('data-src-mp3'):
            if ('British') in span['title']:
                print("Prononciation anglaise : " + span['data-src-mp3'])
            if ('American') in span['title']:
                print("Prononciation américaine : " + span['data-src-mp3'])
```
